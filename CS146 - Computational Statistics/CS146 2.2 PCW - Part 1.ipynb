```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4)) # Create a new figure with a particular size
x = np.linspace(-5, 5, 200) # Array with 200 linearly spaced points in [­5,5]
# Plot 4 normal pdfs with different loc and scale parameters
plt.plot(x, stats.norm.pdf(x, loc=-2), label='loc = -2')
plt.plot(x, stats.norm.pdf(x, loc=2), label='loc = 2')
plt.plot(x, stats.norm.pdf(x, scale=2), label='scale = 2')
plt.plot(x, stats.norm.pdf(x, scale=1/2), label='scale = 1/2')
plt.legend() # Show the legend in the top right corner
plt.show()
```

```python
# Set up the probability distribution
alpha = 2
beta = 1
distribution = stats.gamma(a=alpha, scale=1/beta)
# Plot the pdf
plt.figure(figsize=(8, 4))
x = np.linspace(0, 5)
plt.plot(x, distribution.pdf(x))
plt.title('probability density function')
```

```python
x = np.linspace(-5, 5, 200)

#Normal distribution
plt.plot(x, stats.norm.pdf(x, loc=0, scale= 2), label='loc = 0; scale = 2')
plt.plot(x, stats.norm.pdf(x, loc=-3, scale = 1/2), label='loc = -3;scale = 1/2')
plt.title('Normal distribution')
plt.legend()


# plt.plot(x, stats.norm.pdf(x, scale=2), label='scale = 2')
# plt.plot(x, stats.norm.pdf(x, scale=1/2), label='scale = 1/2')
```

```python
#Gamma distribution

x = np.linspace(0, 5, 200)
plt.plot(x, stats.gamma.pdf(x, a=1, scale= 1/2), label='a=1, scale= 1/2')
plt.plot(x, stats.gamma.pdf(x, a=3, scale = 2), label='a=3, scale = 2')
plt.title('Gamma distribution')
plt.legend()

#np.random.gamma(shape=alpha, scale=scale,size = 100)
```

```python
#Beta distribution

x = np.linspace(0, 5)
plt.plot(x, stats.beta.pdf(x, a=0.5, b= 2), label='a=0.5, b= 2')
plt.plot(x, stats.beta.pdf(x, a=2, b = 1), label='a=2, b = 1')
plt.title('Beta distribution')
plt.legend()
```

```python
# Plot a histogram and pdf of random samples for normal distribution

fig, ax = plt.subplots()
samples = stats.norm.rvs(loc=0, scale= 2,size=100)

# the histogram of the data
n, bins, patches = ax.hist(samples,range=[-5, 5] , normed=True)

# y = ((1 / (np.sqrt(2 * np.pi) * 2)) *
#      np.exp(-0.5 * (1 / 2 * (bins))**2))

x = np.linspace(-5, 5, 11)

#Normal distribution
y= stats.norm.pdf(x, loc=0, scale= 2)

ax.plot(bins, y, '--')

plt.title('Histogram of random samples_normal')
```
