```python
import random
import time

#3 Way Merge Sort

def threewaymergeSort(alist):
    #conditions depending on length of list
    length = len(alist)
    if length < 2:
        return alist
    elif length == 2:
        if alist[0] <= alist[1]:
            return [alist[0], alist[1]]
        else:
            return [alist[1], alist[0]]
    elif length > 2:
        #dividing the list into 3 parts
        left = alist[:((length/3))]
        middle = alist[((length/3)) : ((2*length/3))]
        right= alist[((2*length/3)):]
        
        leftthird = threewaymergeSort(left)
        middlethird = threewaymergeSort(middle)
        rightthird = threewaymergeSort(right)
        
        summation = len(leftthird)+len(middlethird) + len(rightthird)
        sortedlist = [] #empty list for sorting
    
       #initialising counters
        i = 0
        j = 0
        k = 0

        leftthird.append(float("inf")) #sentinels
        middlethird.append(float("inf"))
        rightthird.append(float("inf"))

#appending smallest element to 'sortedlist', to form sorted list in ascending order
        while i < len(leftthird) and j <len(middlethird) and k < len(rightthird) and summation-1>=len(sortedlist):
            if (leftthird[i] <= middlethird[j] and leftthird[i] <=rightthird[k]) :
                sortedlist.append(leftthird[i])
                i+=1
            elif (middlethird[j] <= leftthird[i] and middlethird[j] <= rightthird[k]):
                sortedlist.append(middlethird[j])
                j+=1
            elif (rightthird[k] <= leftthird[i] and rightthird[k] <= middlethird[j]): 
                sortedlist.append(rightthird[k])
                k+=1
        return sortedlist


alist = [random.randint(1,100) for _ in range(100)]    
print "Original list: ",alist 
start = time.time()
print "Sorted List: ", threewaymergeSort(alist)
end = time.time()
print sorted(alist) == threewaymergeSort(alist)
print "Run time for 3-way Mergesort is: " + str(end - start) + " seconds."
```

```python
#insertion sort code
def insertion_sort(a):
    for j in range(1,len(a)):
        key = a[j]
        i = j
        
        while i > 0 and a[i-1] > key:
            a[i] = a[i-1]
            i = i - 1
        a[i] = key
        #replaces the last repeating number
    return a


#three way merge sort with insertion sort

def threewaymergeSortwithInsertionsort(alist, breakPoint): #same as threewaymergesort explanation but with a twist
    length = len(alist)
    if length < 2:
        return alist
    elif length < breakPoint:  #use insertion sort when the length of the sublist is below a specified value.
        return insertion_sort(alist)
    else:
        left = alist[:((length/3))]
        middle = alist[((length/3)) : ((2*length/3))]
        right= alist[((2*length/3)):]
        
        leftthird = threewaymergeSortwithInsertionsort(left, breakPoint)
        middlethird = threewaymergeSortwithInsertionsort(middle, breakPoint)
        rightthird = threewaymergeSortwithInsertionsort(right, breakPoint)

        summation = len(leftthird)+len(middlethird) + len(rightthird)
        sortedlist = []
    
       
        i = 0
        j = 0
        k = 0

        leftthird.append(float("inf"))
        middlethird.append(float("inf"))
        rightthird.append(float("inf"))

        while i < len(leftthird) and j <len(middlethird) and k < len(rightthird) and summation-1>=len(sortedlist):
            if (leftthird[i] <= middlethird[j] and leftthird[i] <=rightthird[k]) :
                sortedlist.append(leftthird[i])
                i+=1
            elif (middlethird[j] <= leftthird[i] and middlethird[j] <= rightthird[k]):
                sortedlist.append(middlethird[j])
                j+=1
            elif (rightthird[k] <= leftthird[i] and rightthird[k] <= middlethird[j]): 
                sortedlist.append(rightthird[k])
                k+=1
        return sortedlist
    
alist = [random.randint(1,100) for _ in range(10)] 

print "Original list: ",alist
start = time.time()
print "Sorted List: ", threewaymergeSortwithInsertionsort(alist, 5)
end = time.time()
print sorted(alist) == threewaymergeSort(alist)
print "Run time for 3-way Mergesort with Insertion Sort is: " + str(end - start) + " seconds."
```
